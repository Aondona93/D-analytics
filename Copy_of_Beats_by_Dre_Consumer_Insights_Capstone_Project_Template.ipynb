## 6. Sentiment Analysis

### 6.1 Methodology
Explain the sentiment analysis techniques used (e.g., TextBlob, NLTK) and provide code snippets.

**Sentiment analysis** involves using natural language processing (NLP) techniques to determine the emotional tone behind a body of text. In this project, various sentiment analysis techniques may be employed, such as:
  1. **TextBlob:**

*  **Overview**: TextBlob is a simple and user-friendly library in Python for processing textual data. It provides an easy API for diving into common natural language processing tasks, including sentiment analysis.
*   **Functionality:**It computes sentiment polarity (a value between -1 and 1, where -1 indicates negative sentiment, 0 is neutral, and 1 is positive) and subjectivity (a value between 0 and 1, where 0 is objective and 1 is subjective).**For example**, a review like “The sound quality is amazing!” would yield a high positive polarity, while “I hate the design” would yield a negative polarity.

2. **NLTK (Natural Language Toolkit):**
*  **Overview:** NLTK is a comprehensive library for natural language processing in Python that provides tools for text processing, classification, tokenization, stemming, tagging, parsing, and more.

*  **Functionality:**NLTK includes various sentiment analysis algorithms, such as the VADER (Valence Aware Dictionary and sEntiment Reasoner) tool, which is specifically designed for sentiment analysis in social media and other informal text.








```python
# preprocess my content column for sentiment analysis to improve accuracy

def preprocess_text(text):
    """Preprocesses text for sentiment analysis."""

    # Convert to lowercase
    text = text.lower()

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join the tokens back into a string
    processed_text = ' '.join(tokens)

    return processed_text

# Apply the preprocessing function to the 'content' column
data_df['processed_content'] = data_df['content'].apply(preprocess_text)

# Print the updated DataFrame
print(data_df[['content', 'processed_content']].head())

# Sample text
text = data_df['processed_content'][0]

# Perform sentiment analysis
blob = TextBlob(text)
sentiment = blob.sentiment
print(sentiment)
```

Present the sentiment analysis results, including visualizations of sentiment distribution, and discuss the overall sentiment towards Beats by Dre products.


```python

# Initialize the sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

# Perform sentiment analysis for each product
product_sentiments = {}
for product in data_df['product_name'].unique():
    product_reviews = data_df[data_df['product_name'] == product]['processed_content'].tolist()
    product_sentiment_scores = []
    for review in product_reviews:
        sentiment = sia.polarity_scores(review)
        product_sentiment_scores.append(sentiment['compound'])
    product_sentiments[product] = product_sentiment_scores

# Analyze the sentiment scores for each product
for product, sentiment_scores in product_sentiments.items():
    avg_sentiment = np.mean(sentiment_scores)
    print(f"Average Sentiment for {product}: {avg_sentiment}")

```

```python
# Ensure the VADER lexicon is downloaded
nltk.download('vader_lexicon')

# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Perform sentiment analysis for each product
product_sentiments = {}
product_avg_sentiment = []  # To store average sentiment per product

for product in data_df['product_name'].unique():
    # Get all reviews for the product based on 'processed_content'
    product_reviews = data_df[data_df['product_name'] == product]['processed_content'].tolist()

    # List to store sentiment scores for the product
    product_sentiment_scores = []

    for review in product_reviews:
        # Get sentiment score for the processed review content
        sentiment = sia.polarity_scores(review)
        # Append the 'compound' sentiment score (overall sentiment)
        product_sentiment_scores.append(sentiment['compound'])

    # Store the sentiment scores for the product
    product_sentiments[product] = product_sentiment_scores

    # Calculate the average sentiment score for each product
    avg_sentiment = np.mean(product_sentiment_scores)
    product_avg_sentiment.append((product, avg_sentiment))

# Convert the data to a DataFrame for easy plotting
product_sentiment_df = pd.DataFrame(product_avg_sentiment, columns=['product_name', 'avg_sentiment'])

# Create a bar plot using Plotly
fig = px.bar(
    product_sentiment_df,
    x='product_name',
    y='avg_sentiment',
    color='avg_sentiment',  # Color scale based on sentiment
    labels={'avg_sentiment': 'Average Sentiment Score', 'product_name': 'Product Name'},
    title="Average Sentiment Score for Each Product"
)

# Customize the layout (optional)
fig.update_layout(
    xaxis_title="Product Name",
    yaxis_title="Average Sentiment Score",
    coloraxis_colorbar=dict(title="Sentiment Score"),
    template="plotly_dark"  # Optional: use a dark theme
)

# Show the plot
fig.show()
```

### 8.2 Comparison of Sentiments
Compare the sentiment analysis results of Beats by Dre with its competitors using visualizations to highlight differences and similarities.

```python
data_df['date_reviewed'] = pd.to_datetime(data_df['date_reviewed'])

# Calculate sentiment polarity and subjectivity using TextBlob
data_df['polarity'] = data_df['processed_content'].apply(lambda text: TextBlob(text).sentiment.polarity)
data_df['subjectivity'] = data_df['processed_content'].apply(lambda text: TextBlob(text).sentiment.subjectivity)

# Define thresholds for categorizing reviews
def categorize_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

def categorize_subjectivity(subjectivity):
    return 'Subjective' if subjectivity >= 0.5 else 'Objective'

# Apply categorization to sentiment polarity and subjectivity
data_df['sentiment_category'] = data_df['polarity'].apply(categorize_sentiment)
data_df['subjectivity_category'] = data_df['subjectivity'].apply(categorize_subjectivity)

# Group by product and sentiment category to get counts
sentiment_counts = data_df.groupby(['product_name', 'sentiment_category']).size().reset_index(name='count')
subjectivity_counts = data_df.groupby(['product_name', 'subjectivity_category']).size().reset_index(name='count')

# Calculate percentages for sentiment categories
total_reviews_by_product = data_df.groupby('product_name').size().reset_index(name='total_reviews')
sentiment_counts = sentiment_counts.merge(total_reviews_by_product, on='product_name')
sentiment_counts['percentage'] = (sentiment_counts['count'] / sentiment_counts['total_reviews']) * 100

# Calculate percentages for subjectivity categories
subjectivity_counts = subjectivity_counts.merge(total_reviews_by_product, on='product_name')
subjectivity_counts['percentage'] = (subjectivity_counts['count'] / subjectivity_counts['total_reviews']) * 100

# 1. Plot for percentage of sentiment categories (Positive, Negative, Neutral)
fig_sentiment = px.bar(
    sentiment_counts,
    x='product_name',
    y='percentage',
    color='sentiment_category',
    barmode='group',
    title='Sentiment Comparison: Beats by Dre vs Competitors (Percentage)',
    labels={'percentage': 'Percentage of Reviews (%)', 'product_name': 'Product Name'},
    text='percentage',
    color_discrete_map={
        'Positive': 'blue',  # Assign blue color for positive sentiment
        'Negative': 'red',   # Assign red color for negative sentiment
        'Neutral': 'grey'    # Assign grey color for neutral sentiment
    }
)

fig_sentiment.update_traces(texttemplate='%{text:.2f}%', textposition='outside')

# Highlight "Beats Pill" label in red by customizing tick labels
custom_sentiment_labels = [
    '<span style="color:red;">Beats Pill</span>' if product == 'Beats Pill' else product
    for product in sentiment_counts['product_name']
]

# Update x-axis tick labels with custom HTML formatting
fig_sentiment.update_layout(xaxis=dict(tickvals=sentiment_counts['product_name'], ticktext=custom_sentiment_labels))

# Show the updated sentiment plot
fig_sentiment.show()

# 2. Plot for percentage of subjectivity categories (Subjective, Objective)
fig_subjectivity = px.bar(
    subjectivity_counts,
    x='product_name',
    y='percentage',
    color='subjectivity_category',
    barmode='group',
    title='Subjectivity Comparison: Beats by Dre vs Competitors (Percentage)',
    labels={'percentage': 'Percentage of Reviews (%)', 'product_name': 'Product Name'},
    text='percentage'
)

fig_subjectivity.update_traces(texttemplate='%{text:.2f}%', textposition='outside')

# Highlight "Beats Pill" label in red by customizing tick labels
custom_subjectivity_labels = [
    '<span style="color:red;">Beats Pill</span>' if product == 'Beats Pill' else product
    for product in subjectivity_counts['product_name']
]

# Update x-axis tick labels with custom HTML formatting
fig_subjectivity.update_layout(xaxis=dict(tickvals=subjectivity_counts['product_name'], ticktext=custom_subjectivity_labels))

# Show the updated subjectivity plot
fig_subjectivity.show()
```
